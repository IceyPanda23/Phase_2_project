```python
# Step 1: Load Libraries
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# Step 2: Load Data
df = pd.read_csv('Led.csv')
df
```

```python
# Testing Hypothesis 1: Education and Economic Development Globally
# Research Question: Is there a relationship between education and economic development on a global scale?
# Null Hypothesis (H0): There is no relationship between education and economic development on a global scale.
# Alternative Hypothesis (H1): There is a relationship between education and economic development on a global scale.

# Step 1: Select variables of interest
df_hypothesis1 = df[['Schooling', 'GDP']].copy()

# Check for missing values and handle them
df_hypothesis1.dropna(inplace=True)
```

```python
# Step 2 :Standardize the variables to handle scaling issues
scaler = StandardScaler()
df_hypothesis1 = scaler.fit_transform(df_hypothesis1)
```

```python
# Step 3: Select the Statistical Test
# We chose the Pearson correlation test to measure the strength and direction of the linear relationship between two continuous variables: years of schooling and GDP per capita.
# The assumptions for this test are:
# - The variables are normally distributed or approximately normal.
# - The variables have a linear relationship.
# - There are no outliers or influential points that distort the relationship.
# - The variables are independent of each other and of other variables.
# We checked these assumptions by plotting histograms, scatter plots, and box plots of the variables and looking for any deviations from normality, linearity, or homoscedasticity (equal variance).
# Conduct the Pearson correlation test
correlation_coefficient, p_value = pearsonr(df_hypothesis1[:, 0], df_hypothesis1[:, 1])
# Print the results
print(f"Correlation coefficient: {correlation_coefficient:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Step 5:setting the significance level
alpha = 0.05

# Make a decision
if p_value < alpha:
    print("\nReject the null hypothesis. There is a relationship between education and economic development on a global scale.")
else:
    print("\nFail to reject the null hypothesis. There is no relationship between education and economic development on a global scale.")
```

```python
# Conduct the Test and Visualize
# Step 6
# Visualize the relationship
plt.figure(figsize=(10, 6))
sns.regplot(x=df_hypothesis1[:, 0], y=df_hypothesis1[:, 1], scatter_kws={'s': 20}, line_kws={'color': 'red'})
plt.title('Scatter Plot and Regression Line: Schooling vs. GDP')
plt.xlabel('Number of Years of Schooling (Standardized)')
plt.ylabel('GDP per Capita (Standardized)')
plt.show()
```

# Hypothesis 1: Education and Economic Development

Research question: Is there a significant correlation between the number of years of schooling and GDP per capita on a global scale?

- Null Hypothesis (H0): There is no significant relationship between the number of years of schooling (Schooling) and Gross Domestic Product per capita (GDP) globally.

- Alternative Hypothesis (H1): There is a significant positive correlation between the number of years of schooling and GDP globally.

- Results: We performed a Pearson correlation test and found that there is a significant positive relationship between schooling and GDP (r = 0.45, p < 0.0001).

## Interpretation:
Based on the correlation coefficient, we can see that there is a moderate positive linear relationship between years of schooling and GDP per capita on a global scale. This means that countries with higher average education levels tend to have higher economic development levels, and vice versa. Based on the p-value, we can see that this relationship is statistically significant at the 0.05 level. This means that the probability of observing such a relationship by chance is very low, and we have enough evidence to reject the null hypothesis and accept the alternative hypothesis.

## Implications:
Our finding suggests that investing in education could be a beneficial strategy for enhancing economic performance and reducing poverty on a global scale. Education may have a positive impact on economic development by increasing human capital, productivity, innovation, and social cohesion.
